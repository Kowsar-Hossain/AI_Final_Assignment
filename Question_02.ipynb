```python
from tensorflow.keras.layers import Input,Dense
from tensorflow.keras.models import Model
def main():
  inputs=Input((3,))
  h1=Dense(5,activation='relu',name="Hidden layer 1")(inputs)
  h2=Dense(6,activation='relu',name="Hidden layer 2")(h1)
  h3=Dense(3,activation='relu',name="Hidden layer 3")(h2)
  outputs=Dense(2,activation='softmax',name="Output layer")(h3)
  model=Model(inputs,outputs)
  model.summary(show_trainable=True)
if __name__=='__main__':
  main()
```
